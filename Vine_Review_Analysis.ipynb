```python
# Import dependencies
import pandas as pd
```

```python
# Read in the vine_table.csv file as a DataFrame
vine_df = pd.read_csv("Resources/vine_table.csv")
vine_df
```

```python
# Filter for all rows where total_votes >= 20
new_df = vine_df.loc[vine_df['total_votes'] >= 20]
new_df
```

```python
# Filter for all rows where helpful_votes divided by total_votes >= 50%
new_new_df = new_df.loc[(new_df['helpful_votes'] / new_df['total_votes']) >= 0.5]
new_new_df
```

```python
# Filter for all rows where a review was written as part of the Vine program (paid)
vine_program_df = new_new_df.loc[new_new_df['vine'] == "Y"]
vine_program_df.head()
```

```python
# Filter for all rows where a review was NOT written as part of the Vine program (paid)
not_vine_program_df = new_new_df.loc[new_new_df['vine'] == "N"]
not_vine_program_df
```

```python
# For Vine Reviews: total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews
vine_total_reviews = vine_program_df['review_id'].nunique()
vine_5star_reviews = vine_program_df.loc[vine_program_df['star_rating'] == 5.0]['review_id'].nunique()
vine_percent_5star_reviews = (vine_5star_reviews / vine_total_reviews) * 100

print(f'Total Paid Reviews: {vine_total_reviews}')
print(f'Total Paid 5 Star Reviews: {vine_5star_reviews}')
print(f'Percent Paid 5 Star Reviews: {vine_percent_5star_reviews}')
```

```python
# For Non-Vine Reviews: total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews
notvine_total_reviews = not_vine_program_df['review_id'].nunique()
notvine_5star_reviews = not_vine_program_df.loc[not_vine_program_df['star_rating'] == 5.0]['review_id'].nunique()
notvine_percent_5star_reviews = (notvine_5star_reviews / notvine_total_reviews) * 100

print(f'Total Unpaid Reviews: {notvine_total_reviews}')
print(f'Total Unpaid 5 Star Reviews: {notvine_5star_reviews}')
print(f'Percent Unpaid 5 Star Reviews: {notvine_percent_5star_reviews}')
```
